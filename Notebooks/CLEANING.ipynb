### Churn Target Changing To Binary

* Replace True with 1 
 
* Replace False with 0

```python
df['churn'] = df['churn'].apply(lambda x: 1 if x== True else 0)
```

```python
df
```

### International Plan Binary

- Replace Yes w/1
- Replace No w/0
- Need to change from string to int

```python
df['international_plan'].replace(to_replace = 'yes', value='1',inplace=True)
```

```python
df['international_plan'].replace(to_replace = 'no', value='0',inplace=True)
```

```python
df['international_plan'] = df['international_plan'].astype(int)
```

```python
df['international_plan'].value_counts()
```

### Voicemail Plan Binary

- Replace Yes w/1
- Replace No w/0
- Again need to switch from str to int

```python
df['voice_mail_plan'].replace(to_replace = 'yes', value='1',inplace=True)
```

```python
df['voice_mail_plan'].replace(to_replace = 'no', value='0',inplace=True)
df['voice_mail_plan'] = df['voice_mail_plan'].astype(int)
```

```python
df.info()
```

```python
df['total_calls'] = df['total_day_calls']+ df['total_eve_calls']+ df['total_night_calls']+ df['total_intl_calls']
```
